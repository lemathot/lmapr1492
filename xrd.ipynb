```python
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifWriter
from pymatgen.analysis.diffraction.xrd import * 
from matplotlib import pyplot as plt 
from tabulate import tabulate
```

```python
with MPRester("SsRPm7BKULyAwhtl") as m:

    structure = m.get_structure_by_material_id("mp-7476")
```

# Diffraction des rayons X du RbDyO2

Nous allons déterminer les trois premiers pics du diffractogramme du RbDyO2 pour un rayonnement X incident Cu Ka de longueur d'onde λ=1.54060 Å.

```python
xrd = XRDCalculator('CuKa')
patt = xrd.get_pattern(structure)
```

```python
#Premier pic
print("Premier pic du diffractogramme pour un rayonnement X incident CuKa de longueur d’onde λ=1.54060 Å")
print("-------------------------------------------------------------------------------------------------")
print(" ")
print("Angle 2\u0398 [°] :", patt.x[0])
print("Intensité du pic [%] :", patt.y[0])
print("Indices hkl associés :", patt.hkls[0])
print("Distance interplanaire d :", patt.d_hkls[0])
```

```python
#Deuxième pic
print("Deuxième pic du diffractogramme pour un rayonnement X incident CuKa de longueur d’onde λ=1.54060 Å")
print("--------------------------------------------------------------------------------------------------")
print(" ")
print("Angle 2\u0398 [°] :", patt.x[1])
print("Intensité du pic [%] :", patt.y[1])
print("Indices hkl associés :", patt.hkls[1])
print("Distance interplanaire d :", patt.d_hkls[1])
```

```python
#Troisième pic
print("Troisième pic du diffractogramme pour un rayonnement X incident CuKa de longueur d’onde λ=1.54060 Å")
print("---------------------------------------------------------------------------------------------------")
print(" ")
print("Angle 2\u0398 [°] :", patt.x[2])
print("Intensité du pic [%] :", patt.y[2])
print("Indices hkl associés :", patt.hkls[2])
print("Distance interplanaire d :", patt.d_hkls[2])
```
